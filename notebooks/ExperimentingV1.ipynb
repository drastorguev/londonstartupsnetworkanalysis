```python
import networkx as nx
import json
import matplotlib.pyplot as plt
%matplotlib inline
G = nx.erdos_renyi_graph(50,4.0/50)
while not nx.is_connected(G):
    G = nx.erdos_renyi_graph(30,4.0/30)
plt.figure(figsize=(10,4));
nx.draw(G)
```

```python
Z=nx.Graph()
Z.add_node("a")
Z.add_nodes_from(["b","c"])
Z.add_edges_from([("a","c"),("c","d"), ("a",1), (1,"d"), ("a",2)])
print("Nodes of graph: ")
print(Z.nodes())
print("Edges of graph: ")
print(Z.edges())
nx.draw(Z,with_labels=True)
plt.show() # display
```

```python
A=nx.Graph()
A.add_edges_from([("1","2"),("2","3"), ("1","3"), ("3","4"),("4","5"),("7","5"),("4","1"),("1","9"),("6","4"),("2","5"),("3","6"),("5","9")])
nx.draw(A,with_labels=True)
plt.show() # display
```
