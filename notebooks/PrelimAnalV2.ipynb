```python
import pickle
import json
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

V3=nx.Graph()

targets = ['swimgeek','PshyMorphic','phpeach', 'colinhayhurst','jibly']

for eachtarget in targets:
    savedfollowers = pickle.load( open( "../pickles/"+eachtarget+".p", "rb" ) )
    for eachname in savedfollowers:
        V3.add_edge(eachtarget,eachname)
nx.info(V3)
```

```python
from networkx.readwrite import json_graph
d3json = json_graph.node_link_data(V3)
d3json['links'] = [{'source': d3json['nodes'][link['source']]['id'], 'target': d3json['nodes'][link['target']]['id']} for link in d3json['links']]
targetjson = json.loads(json.dumps(d3json))
del targetjson["directed"]
del targetjson["graph"]
del targetjson["multigraph"]
with open('visualdatav2.json', 'w') as f:
     json.dump(targetjson, f)
print targetjson
```
