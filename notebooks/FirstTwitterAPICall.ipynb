```python
import requests
from requests_oauthlib import OAuth1
import json
from details import *

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

V1=nx.Graph()

url = 'https://api.twitter.com/1.1/account/verify_credentials.json'
auth = OAuth1(item1, item2, item3, item4)
requests.get(url, auth=auth)

twitterprofile = ['dsrastorguev','rsjhayes','willgdjones']
for eachprofile in twitterprofile:
    
    initialrequest = requests.get('https://api.twitter.com/1.1/followers/list.json?cursor=-1&screen_name='+eachprofile+'&skip_status=true&include_user_entities=false&count=200', auth=auth)
    for eachtuser in initialrequest.json()['users']:
        V1.add_edge(eachprofile,eachtuser['screen_name']) 
    
nx.draw(V1,with_labels=True,node_size=30,font_size=5)
plt.figure(figsize=(50,50))
plt.show() # display
    
```

```python
V2=nx.Graph()

twitterprofile2 = ['fb_engineering','TwitterEng','SpotifyEng']
for eachprofile in twitterprofile2:
    initialrequest = requests.get('https://api.twitter.com/1.1/followers/ids.json?cursor=-1&screen_name='+eachprofile+'&count=500', auth=auth)
    for eachtuser in initialrequest.json()['ids']:
        V2.add_edge(eachprofile,eachtuser) 
nx.draw(V2)
plt.figure(figsize=(100,100))
plt.show() # display
```
