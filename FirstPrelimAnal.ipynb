```python
import pickle

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

V3=nx.Graph()

targets = ['swimgeek','PshyMorphic','phpeach']

for eachtarget in targets:
    savedfollowers = pickle.load( open( eachtarget+".p", "rb" ) )
    for eachname in savedfollowers:
        V3.add_edge(eachtarget,eachname)
        
nx.draw(V3,with_labels=True,node_size=15,font_size=4)
plt.figure(figsize=(500,500))
plt.show() # display
nx.info(V3)
```

```python
from networkx.readwrite import json_graph

testing = json_graph.node_link_data(V3)
```

```python
testing['links'] = [{'source': testing['nodes'][link['source']]['id'], 'target': testing['nodes'][link['target']]['id']} for link in testing['links']]
```

```python
import json
targetjson = json.loads(json.dumps(testing))
del targetjson["directed"]
del targetjson["graph"]
del targetjson["multigraph"]
print json.dumps(targetjson, indent=4)
```
